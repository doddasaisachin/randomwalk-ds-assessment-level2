### Function to print mean,median and mode of columns

```python
def print_central_tend(dataset,col):
    print('mean :',dataset[col].mean())
    print('median:',dataset[col].median())
    print('mode:',dataset[col].mode()[0])
```

### Below function plots the KDE distribution of a given column from a dataset and also plot the vertical lines of the mean, median, and mode of respective column.

```python
def func(dataset,col):
    sns.kdeplot(dataset[col])
    sns.despine(top=True, right=True, left=True)
    mean_val = dataset[col].mean()
    median_val = dataset[col].median()
    mode_val = dataset[col].mode()[0]

    plt.axvline(mean_val, label="Mean")
    plt.axvline(median_val,c='0',label="Median")
    plt.axvline(mode_val,c='r',label="Mode")

    plt.legend()
```

# Adelie Penguin

### I) bill_depth_mm

```python
func(df_Adelie,'bill_depth_mm')
```

SKEWNESS: As 'mean < median < mode' the data is negatively skewed that is LEFT-SKEWED.

KURTOSIS: As the above plot shows peak close to center and with moderate tails so this is MESOKURTIC

### II) bill_length_mm

```python
func(df_Adelie,'bill_length_mm')
```

```python
print_central_tend(df_Adelie,'bill_length_mm')
```

SKEWNESS: 

As mean and median are almost same but still mode is higher than those two there is a slight           RIGHT-SKEWNESS

KURTOSIS: As the above plot has good peak and normal tail this is MESOKURTIC
