```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Preprocessed_df.csv")
```

```python
df.head()
```

```python
num_cols=[]
```

```python
for i in df.dtypes.index[1:]:
    if df[i].dtype!='O':
        num_cols.append(i)
```

```python
num_cols
```

```python
for i in num_cols:
    sns.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```

From the above plot First three features data doesn't contain any outliers

'body_mass_g' is the only features which having outliers
