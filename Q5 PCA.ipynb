The curse of dimensionality can be determined by:
    
    
Features contribute very little to the variance (low explained variance ratio).

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df[['bill_length_mm', 'bill_depth_mm']])
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```
