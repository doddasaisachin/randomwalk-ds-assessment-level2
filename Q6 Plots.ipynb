```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Preprocessed_df.csv')
```

```python
df.head()
```

```python
X='bill_length_mm'
Y='bill_depth_mm'
```

# 1. Scatter Plot

```python
sns.scatterplot(x=df[X],y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

# 2. pairplot

```python
sns.pairplot(df[['bill_length_mm', 'bill_depth_mm']])
plt.show()
```

# 3. LinePlot

```python
sns.lineplot(x=df[X], y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

# 4. BarPlot

```python
sns.barplot(x=df[X], y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

# 5. BoxPlot

```python
sns.boxplot(x=df[X], y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

# 6. ViolinPlot

```python
sns.violinplot(x=df[X], y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

# 7. HeatMap

```python
sns.heatmap(df[['bill_length_mm', 'bill_depth_mm']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

# 8. Regression Plot

```python
sns.regplot(x=df[X], y=df[Y])
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```
